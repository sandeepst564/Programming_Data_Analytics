```python
#GCD and LCM
a = int(input("Enter first number"))
b = int(input("Enter seconf number"))
x,y = a,b
while y:
    x,y= y,x%y
    gcd = x
    lcm= (a*b) // gcd
    print("GCD:",gcd)
    print("LCM",lcm)
```

```python
#Oldest Person
people = {
    "Sandeep":29,
    "Varun": 20,
    "Himanshu": 55
}
oldest = max(people, key=people.get)
print(oldest,people[oldest])
```

```python
#Bubble Sort
arr = [5,7,8,4]
for i in range (len(arr)):
    for j in range (len(arr)-1):
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1] = arr[j+1],arr[j]
print(arr)
```

```python
#Remove Duplicates
list = [22,22,56,78,3,3,4]
new_list = []
for x in list:
    if x not in new_list:
        new_list.append(x)
print(new_list)
```

```python
#Prime Numbers (1–50)
for num in range(2, 51):
    prime = True 
    for i in range(2, num):
        if num % i == 0:
            prime = False
            break
        if prime:
            print(num)
```

```python
#Number Guessing Game
import random
num = random.randint(1,100)
for i in range (10):
    try:
        guess = int (input("Guess the number:"))
        if guess == num:
           print("guess is correct")
           break
        elif guess > num:
           print("Very high")
        else:
           print("Very low")
    except ValueError:
        print("Enter the vaild number",num)
else:
    print("Number was",num)
```

```python
#Sum and Average of List
num = [20,99,75,34]
total = 0
for n in num:
    total += n 
    average = total / len(num)
    print("Sum :", total)
    print("Average :", average)
```

```python
#Temperature Converter (C, F, K)
temp = float(input("Enter temperature:"))
unit = input("Enter unit:")
if unit == "C":
    print("Fahrenheit:", (temp * 9/5)+32)
    print("kelvin:", temp+273.15)
elif unit =="F":
    c = (temp - 32)* 5/9
    print("Celsius:", c)
    print("Kelvin:", c + 273.15)
elif unit == "K":
    c = temp - 273.15
    print("Celsius:", c)
    print("Fahrenheit:", (c * 9/5) + 32)

else:
    print("Invalid unit")
```
