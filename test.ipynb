```python
#Rectangle Area (Input Validation – Simple)
length = float(input("Enter length for Rectangle Area: "))
width = float(input("Enter width for Rectangle Area: "))

if length > 0 and width > 0:
    area = length * width
    print("Area of rectangle:", area)
else:
    print("Please enter positive numbers only")
```

```python
#Temperature Converter (C, F, K)
temp = float(input("Enter temperature:"))
unit = input("Enter unit:")

if unit == "C":
    print("Fahrenheit:", (temp * 9/5) + 32)
    print("Kelvin:", temp + 273.15)

elif unit == "F":
    c = (temp - 32) * 5/9
    print("Celsius:", c)
    print("Kelvin:", c + 273.15)

elif unit == "K":
    c = temp - 273.15
    print("Celsius:", c)
    print("Fahrenheit:", (c * 9/5) + 32)

else:
    print("Invalid unit")
```

```python
#Sum and Average of List
numbers = [10, 20, 30, 40, 50]

total = 0
for n in numbers:
    total += n
average = total / len(numbers)
print("Sum:", total)
print("Average:", average)
```

```python
#Even or Odd
num = int(input("Enter a number:"))
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
#Number Guessing Game
import random
number = random.randint(1, 100)
for i in range(5):
    guess = int(input("Guess the number: "))

    if guess == number:
        print("You guessed it right!")
        break
    elif guess > number:
        print("Too high")
    else:
        print("Too low")
print("Number was:", number)
```

```python
#Palindrome Checker
text = input("Enter text: ")
text = text.lower().replace(" ", "")
if text == text[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
#Factorial (Iterative)
n = int(input("Enter a number: "))
fact = 1
for i in range(1, n + 1):
    fact = fact * i

print("Factorial:", fact)
```

```python
#Largest Number in List
numbers = [12, 45, 7, 89, 23]
largest = numbers[0]
for n in numbers:
    if n > largest:
        largest = n
print("Largest number:", largest)
```

```python
#Simple Calculator
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
operator = input("Enter operator (+ - * /): ")
if operator == "+":
    print(a + b) #addition
elif operator == "-":
    print(a - b)#subtraction
elif operator == "*":
    print(a * b)#multiplication
elif operator == "/":
    print(a / b)#division
else:
    print("Invalid operator")
```

```python
#Square Root (Simple Loop)
num = float(input("Enter a number: "))
guess = num / 2
for i in range(10):
    guess = (guess + num / guess) / 2
print("Square root:", guess)
```

```python
#Reverse a List
lst = [1, 2, 3, 4, 5]
rev = []
for i in lst:
    rev.insert(0, i)
print("Reversed list:", rev)
```

```python
#Prime Numbers (1–50)
for num in range(2, 51):
    prime = True
    for i in range(2, num):
        if num % i == 0:
            prime = False
            break
    if prime:
        print(num)
```

```python
#Fibonacci Series
n = int(input("Enter limit: "))
a = 0
b = 1
while a <= n:
    print(a)
    a, b = b, a + b
```

```python
#Vowel Counter
text = input("Enter text: ").lower()
count = 0
for ch in text:
    if ch in "aeiou":
        count += 1
print("Vowels:", count)
```

```python
#Leap Year
year = int(input("Enter year: "))
if year % 400 == 0 or (year % 4 == 0 and year % 100 != 0):
    print("Leap year")
else:
    print("Not a leap year")
```

```python
#Remove Duplicates
lst = [1, 2, 2, 3, 1, 4]
new_list = []
for x in lst:
    if x not in new_list:
        new_list.append(x)
print(new_list)
```

```python
#Bubble Sort
arr = [5, 3, 8, 4, 2]
for i in range(len(arr)):
    for j in range(len(arr) - 1):
        if arr[j] > arr[j + 1]:
            arr[j], arr[j + 1] = arr[j + 1], arr[j]
print(arr)
```

```python
#Oldest Person
people = {"Alice": 30, "Bob": 25, "Charlie": 40}
oldest = max(people, key=people.get)
print(oldest, people[oldest])
```

```python
#GCD and LCM
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
x, y = a, b
while y:
    x, y = y, x % y
gcd = x
lcm = (a * b) // gcd
print("GCD:", gcd)
print("LCM:", lcm)
```
